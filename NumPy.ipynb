# Data Manipulation with NumPy


**NumPy** (Numerical Python) is one of the most fundamental and powerful libraries for numerical computing in Python. It provides support for large multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays.

In this section, we will explore the key concepts of NumPy, and learn how to work with arrays and perform various mathematical operations.

### What is NumPy?

NumPy is a Python library that offers:
- Efficient handling of large arrays and matrices.
- A wide range of mathematical, logical, and statistical functions.
- Support for linear algebra, Fourier transforms, random number generation, and more.

## Concatenate

```python
arr1 = np.arange(1, 10).reshape((3, 3))
arr2 = np.arange(10, 19).reshape((3, 3))
print(arr1)
print(arr2)
```

```python
arr1 + arr2
```

```python
arr = np.concatenate((arr1, arr2), axis=0)
print(arr)
arr.shape
```

```python
arr = np.concatenate((arr1, arr2), axis=1)
print(arr)
arr.shape
```

```python
arr2 = arr2[2, ::].reshape((1, 3))
arr2
```

```python
arr = np.concatenate((arr1, arr2), axis=0)
print(arr)
arr.shape
```
