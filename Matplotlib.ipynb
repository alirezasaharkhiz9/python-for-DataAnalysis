# Data Visualization with Matplotlib

---------------

## Introduction to Matplotlib

**Matplotlib** is one of the most widely used libraries for data visualization in Python. It provides an object-oriented API for embedding plots into applications and a high-level interface for creating static, animated, and interactive visualizations. **Matplotlib** is especially known for its ability to create a wide range of plots like line charts, bar charts, histograms, scatter plots, and more.

The most commonly used part of **Matplotlib** is its `pyplot` module, which is typically imported as `plt` and offers functions that make it easy to generate plots with just a few lines of code.

### Why Use Matplotlib?

1. **Comprehensive Visualization**:
   - **Matplotlib** allows you to create various types of visualizations such as line plots, histograms, pie charts, scatter plots, etc.
   
2. **Customization**:
   - It offers a high level of control over plot customization, such as labels, titles, legends, colors, and much more.

3. **Integration**:
   - **Matplotlib** works well with other data science libraries like **Pandas** and **NumPy**, making it a great tool for visualizing and analyzing data.

4. **Interactive Plots**:
   - With **Matplotlib**, you can also create interactive visualizations that can be embedded into applications or websites.

## Import matplotlib

## Basic Plots

### plt.plot()

```python
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(10, 100, 50)
x.shape
```

```python
y = np.linspace(40, 80, 50)
y.shape
```

```python
plt.plot(x, np.sin(y))
plt.show()
```

### xlabel, ylabel

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### xlim, ylim

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 200)
plt.ylim(0, 200)
plt.show()
```

### legend, title

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 200)
plt.ylim(0, 200)
plt.title('plot')
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 200)
plt.ylim(0, 200)
plt.title('plot')
plt.legend(['plot'])
plt.show()
```

### plt.figure()

fig.add_subplot()

fig.add_axes()

```python
fg = plt.figure()

ax1 = fg.add_axes((0, 0, 0.4, 1))
ax1.plot(x, y)
plt.show()
```

```python
fg = plt.figure(figsize=(2, 2))

ax1 = fg.add_axes((0, 0, 0.4, 1))
ax1.plot(x, y)

ax2 = fg.add_axes((1, 1, 0.5, 0.8))
ax2.plot(x, y)

plt.show()
```

```python
fg = plt.figure()

ax1 = fg.add_subplot(1, 2, 1)
ax1.plot(x, y)

ax1 = fg.add_subplot(1, 2, 2)
ax1.plot(x, y)

plt.show()
```

### plt.hist()

### plt.scatter()

```python
plt.scatter(data['lstat'], data['medv'])
plt.show()
```

### plt.bar()

```python
fruits = ['apple', 'banana', 'orange', 'grape', 'kiwi']
counts = [10, 15, 7, 5, 12]
```

```python
plt.bar(fruits, counts)
plt.show()
```

```python
uni_val, counts = np.unique(data['chas'], return_counts=True)

plt.bar(uni_val, counts, tick_label=['0', '1'])
plt.show()
```

### plt.pie()

```python
plt.pie(counts, labels=fruits)
plt.show()
```
